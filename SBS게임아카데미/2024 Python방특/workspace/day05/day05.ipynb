```python
#set 안에 굳이 set을 넣고싶을 때 쓰는 방법
#frozenset
#frozenset으로 만든 집합은 내부 요소를 바꿀 수 "없다".
a = frozenset(range(10))
print(a)
print(set(a))

b = set(range(10))
print(b)
print(frozenset(b))
```

```python
#합집합(UNION)
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print(a,b)
print(set.union(a,b))
c = a | b # or을 사용해도 합집합
print(c)

# 교집합 (INTERSECTION)
print(set.intersection(a,b))
d = a & b # and를 사용하면 교집합
print(d)

#차집합 (difference) -> 순서에 따라 다르다
print(set.difference(a,b))
print(set.difference(b,a))

#대칭차집합 (symmetric_difference) -> 합집합 - 교집합
print(set.symmetric_difference(a,b))
e = a ^ b
```

```python
a = {1, 2, 3, 4,}
print(a)
a |={5}  # 세트에 요소 추가하기, 집합 |= 다른집합
print(a)
a.update({6}) # update로도 요소 추가 가능
print(a)
a &= {3, 4, 5, 6, 7, 8} # 집합.intersection_update(다른 집합) 과 같은 기능
print(a)
a -= {4,5} # 요소 제거, 집합.difference_update(다른 집합)
print(a)
a ^= {5, 6} # 집합.symmetric_difference_update(다른 집합)
print(a)
```

```python
# 부분 집합
a = {3, 4, 5}
b = {1, 2, 3, 4, 5, 6}
print(a <= b) # True가 나온다면 집합 a는 집합 b에 속함을 나타낸다
print(a.issubset(b))

# 진부분 집합
print(a < b)

# 상위 집합
print(a >= b)
print(a.issuperset(b))

# 진상위집합
print(a > b)
```

```python
# 비교, 두 집합의 구성이 같은지
a = {1, 2, 3}
b = {3, 2, 1}
print(a == b) #같냐
print(a != b) #다르냐
```

```python
# 교집합이 없냐 확인
a = {1, 2, 3 ,4}
print(a.isdisjoint({4,5}))
print(a.isdisjoint({5,6}))
```

```python
# 집합에 요소 추가
a = {1, 2 ,3}
print(a)
a.add(4)
print(a)
a.add(3) # 기존 요소를 추가해도 에러는 안남 결과는 기존과 동일
print(a)

# 집합의 요소 제거 3가지
a.remove(3)
print(a)
# a.remove(5) 없는 요소를 제거하려 하면 오류 발생 keyError
a.discard(2)
print(a)
a.discard(5) # 이건 오류가 나진 않음 없는 요소면 그냥 넘어간다
print(a)
print(a.pop()) # 임의의 요소를 가져와 출력하고 삭제한다, 요소가 없으면 error
print(a)
a.clear() # 모든 요소 삭제
print(a)
```

```python
# 99페이지 4번
# 마지막 차량번호 4자리를 더했을 때
# + 기호 사용하지 않고 합 구하기
# 짝수로 끝나면 -> 운행가능
# 홀수로 끝나면 -> 운행불가
# 237가12345 -> 10 -> 운행가능

car = input("차량 번호를 입력하세요 >>> ")
a = list(car)
b = a[-4:]   # 이 두 줄을 a = list(car[-4:]) 로 한 번에 가능
print(b)
print(type(b))

c = list(map(int,b))
print(c)

if sum(c) % 2 == 0:
  print("운행가능")
else:
  print("운행불가")

```

```python
# 평어 A+, A0, A-, B+, B0, B-, C+, C0, C-, D+, D0, D-
#     4.3 4.0 3.7 3.3 3.0 2.7 2.3 2.0 1.7 1.3 1.0 0.7
a = input("평어 입력 >>> ")
print(a[0])
print(a[1])

if a[0] == 'A':
  b = 4
elif a[0] == 'B':
  b = 3
elif a[0] == 'C':
  b = 2
elif a[0] == 'D':
  b = 1

if a[1] == '+':
  b += 0.3
elif a[1] == '-':
  b -= 0.3

print(b)
```

```python
# 평어 A+, A0,  A-,  B+,  B0,  B-,  C+,  C0,  C-,  D+,  D0,  D-
#     4.3  4.0  3.7  3.3  3.0  2.7  2.3  2.0  1.7  1.3  1.0  0.7
# dictionary로 풀기
a = input("평어 입력 >>> ")
print(a[0])
print(a[1])
dic1 = dict(zip('ABCD',[4, 3, 2, 1]))
dic2 = dict(zip('+0-',[0.3, 0.0, -0.3]))

b = dic1[a[0]] + dic2[a[1]]
print(b)
```
