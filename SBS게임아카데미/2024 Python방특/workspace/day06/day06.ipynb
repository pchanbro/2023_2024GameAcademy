```python
# 입력할 수의 개수를 입력 >> 3
# 수1을 입력 >>> 100
# 수2를 입력 >>> 200
# 수3을 입력 >>> 300
# 세 개의 수를 리스트에 넣고
# 리스트를 뒤집은 결과는 300 200 100 이 출력하도록 만들기
num = int(input("입력할 수의 개수를 입력하세요 >>> "))

a = list()

for i in range(1, num + 1):
  b = int(input(f"수 {i}를(을) 입력하세요 >>> "))
  a.append(b)

print(a)

print("리스트를 뒤집은 결과는")
for i in range(2, -1, -1):
  print(a[i], end = ' ')

print()
# 선생님거
for i in reversed(range(num)):
  print(a[i], end = ' ')
```

```python
# 문자열을 입력 >>> Python
# step값을 입력 >>> 2
# 리스를 출력한 결과
# {'p', 't', 'o'}

str1 = input("문자열을 입력하세요 >>> ")
step = int(input("step값을 입력하세요 >>> "))

slist = []

for i in range(0, len(str1), step):
  slist.append(str1[i])

print("리스트를 출력한 결과", slist)
```

```python
# 113p 6번
# 구구단 출력
# 2*1=2 3*1=3 .... 9*1=9
# ...
# 2*9=18 .... 9*9=81
# 바깥은 while 안쪽은 for
a = 1
while a < 10:
  for i in range(2,10):
    print(i,"*",a,"=",a*i, end = '\t') #print(f"{i} * {a} = {a*i}", end = '\t') 이렇게 해도 됨
  a += 1
  print()
```

```python
# 130 페이지 4번
exam = [99, 78, 100, 91, 81, 85, 54, 100, 71, 50]
#각 점수에 대해서 5점씩 더하는데, 더한 값이 100 넘지 않도록

#출력 결과
#[100, 83, 100, 96, 86, 90, 59, 100, 76, 55]
for i in range(len(exam)):
  exam[i] += 5
  if exam[i] > 100:
    exam[i] = 100

print(exam)
```

```python
# 130페이지 5번
# for문 1개 이용해서
# 1부터 99까지의 수를 대상으로 369 게임의 결과 출력하기
# 1 2 짝 4 5 짝 ...

for i in range(1, 100):
  if i % 10 == 3 or i % 10 == 6 or i % 10 == 9 or i // 10 == 3 or i // 10 == 6 or i // 10 == 9:
    if (i % 10 == 3 or i % 10 == 6 or i % 10 == 9) and (i // 10 == 3 or i // 10 == 6 or i // 10 == 9):
      print("짝짝", end = '\t')
    else:
      print("짝" , end = '\t')
    if i % 10 == 0:
      print()
    continue
  print(i, end = '\t')
  if i % 10 == 0:
    print()
```

```python
# 141페이지 1번
# 현재 10000원을 갖고 있습니다. 얼마를 사용할 것인지 반복해서 입력받아 10000원을 모두 사용하세요
# 0 이하의 금액은 사용할 수 없으며, 현재 가지고 있는 돈보다 더 큰 금액도 사용할 수 없습니다.

money = 10000

while money > 0:
  print(f"남은 금액 : {money}")
  use = int(input("사용할 금액을 말씀하세요 >>> "))
  if use > money:
    print("{0}원이 부족합니다, 소지한 금액 이상의 돈을 사용할 수 없습니다.".format(use - money))
    continue
  elif use <= 0:
    print("0원 이하의 금액은 사용할 수 없습니다.")
    continue
  else:
    money -= use
    print(f"{use}원 사용했습니다.")

print("모든 돈을 사용했습니다.")
```
