```python
f = open('level.txt', 'r')
s = f.read()
a = str(int(s) + 1)
with open('level2.txt.', 'w') as g:
  g.write(a + '\n') # 써줄 때 줄바꿈을 하려면 \n을 써줘야 한다.
  g.write(a)
f.close()
```

```python
# 리스트에 있는 문자열들 한 번에 추가하기 writelines 함수 사용
fruit = ['apple\n', 'orange\n', 'banana\n'] # 이 경우엔 리스트 안의 글자들에 \n을 추가해줘야 줄바꿈이 된다.

with open('fruit.txt', 'w') as f:
  f.writelines(fruit)
```

```python
with open('fruit.txt', 'r') as f:
  lines = f.readlines() # 파일의 내용을 한 번에 한 줄씩 여러 번 가져오기
  print(lines)
```

```python
with open('fruit.txt', 'r') as f:
  line = None
  while line != '': # 파일에 문자열이 몇 개 있을지 모르기 때문에 while 반복문을 이용해 계속 받음
    line = f.readline()
    print(line.rstrip('\n')) #readline()은 \n까지 가져오기 때문에(바로 위 참고) 출력할 때는 rstrip('\n')을 해야함
```

```python
# for문에 파일 객체를 넣어 반복문을 돌려도 된다. 파일 객체에서 파일의 내용을 한 줄씩 꺼내와서 변수에 저장
with open('fruit.txt', 'r') as f:
  for line in f:
    print(line.rstrip('\n'))
```

```python
# 파일에 있는 글자를 가져와 리스트로 만드는 방법
with open('fruit.txt', 'r') as f:
  g = list(map(lambda x : x.rstrip('\n') ,list(f))) # map으로 할 때는 rstrip을 사용할 수 없어서 lambda를 사용해야 함
  print(g)
```

```python
# 피클링 (pickling, 객체 저장하기)
import pickle

name = 'james'
age = 20
address = '서울 강남구'
score = {'python':90, 'java':80, 'ai':70}

with open('james.p', 'wb') as f: # 피클을 쓸땐 wb, 읽을 때 rb
  pickle.dump(name, f)
  pickle.dump(age, f)
  pickle.dump(address, f)
  pickle.dump(score, f)
```

```python
# 언피클링 (객체 읽어오기)
import pickle

with open('james.p', 'rb') as f:
  name = pickle.load(f)
  age = pickle.load(f)
  address = pickle.load(f)
  score = pickle.load(f)
  print(name)
  print(age)
  print(address)
  print(score)
```

```python
# 추가 파일 모드
# r, w -> 텍스트 읽기, 초기화 하고 쓰기
# rb, wb -> 바이너리 읽기, 초기화 하고 쓰기
# a -> 파일에 추가하기
# x -> 파일이 있으면 쓸 수 없다고 오류 발생, 파일이 없으면 파일을 생성하고 파일 작성, 배타적 생성(쓰기)모드
```

```python
fruit = ['apple', 'banana', 'orange', 'mango', 'grape', 'blackberry']

with open('fruits.txt', 'w') as f:
  f.writelines(list(map(lambda x : x+'\n', fruit))) # lambda를 통해 리스트 요소 각각에 \n을 적어주지 않고도 추가하도록 해줌

with open('fruits.txt', 'r') as f:
  g = list(map(lambda x : x.rstrip('\n') ,list(f)))
  h = []
  for i in g:
    if len(i) == 5:
      h.append(i)
  h.sort(reverse = True)

  print(*h, sep = '\n') # 언패킹 사용됨 뭔지 찾아보기
```

```python
import pickle
with open('account.p', 'wb') as f:
  pickle.dump({},f)
```

```python
while True:
  mode = input("모드를 입력하세요(회원가입, 로그인) >>> ")
  if mode == '회원가입':
    id = input("아이디 >>> ")
    password = input("비밀번호 >>> ")

    account = {}
    f = open('account.p', 'rb')
    account = pickle.load(f)
    if id in account:
      print("존재하는 아이디입니다.")
      f.close()
    else:
      f.close()
      with open('account.p', 'wb') as f:
        account[id] = password
        pickle.dump(account,f)
        print("회원가입이 완료되었습니다.")

  elif mode == '로그인':
    id = input("아이디 >>> ")
    password == input("비밀번호 >>> ")

    account = {}
    with open('account.p', 'rb') as f:
      account = pickle.load(f)

      if id in account:
        if account[id] == password:
          print(f'{id}님 환영합니다.')
        else:
          print("비밀번호가 틀렸습니다.")
      else:
        print("존재하지 않는 계정입니다.")
  else:
    print("모드를 다시 입력하세요.")
```

```python
# 독스트링(docstring (document + string)) 주석을 쓰지 않고 문자열(""" """)을 사용
def function1(num1, num2):
  """ 이 함수는 num1과 num2의 곱을 출력 """
  print(num1 * num2)

function1(2,3)
print(function1.__doc__)
print(help(function1))
```

```python
def function2(num1, num2):
  """
  함수 이름 : function2
  함수 요약 : 두 수를 곱하고, 첫 번째 매개변수를 더한다.

  매개변수
  - num1 : 정수 값, 첫 번째 수
  - num2 : 정수 값, 두 번째 수

  반환 값
  - 정수를 반환, 두 매개변수를 연산하여 반환

  상세 설명
  - 두 수를 곱하고, 첫 번째 매개변수를 더한다
  """
  return num1 + num2 * num1

print(function2(5,7))
print(function2.__doc__)
```

```python
def repeatString(str1, num1):
  str2 = ''
  num1 = int(num1)

  for i in str1:
    for j in range(num1):
      str2 += i

  return str2

str1 = input("반복할 문자열을 입력하세요 >>> ")
num1 = input("반복할 횟수를 입력하세요 >>> ")
output = repeatString(str1, num1)
print(output)
```
